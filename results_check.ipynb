```python
import pandas as pd
from functions.model import Player, Game, session
```

```python
players = session.query(Player).all()
players_data = []
for player in players:
    players_data.append({
        'ID': player.id,
        'Name': player.first_name,
        'Surname': player.last_name,
        'Height inches': player.height_inches,
        'Height feet': player.height_feet
    })

df_player = pd.DataFrame(players_data)

df_player
```

```python
Amir_id = df_player.loc[df_player['Name']=='Amir', 'ID'].values[0]
Amir_id
```
